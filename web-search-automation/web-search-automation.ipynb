# Web Automation and Information Retrieval with Python

#### Challenge: 

We work in an import company and the price of our products is linked to the quotation of:
- Dollar
- Euro
- Gold

We need to get on the internet, in an automatic way, the quotation of these 3 items and know how much we should charge for our products, considering a contribution margin that we have in our database.

Data Base: https://drive.google.com/drive/folders/1KmAdo593nD8J9QBaZxPOG1yxHZua4Rtv?usp=sharing

To do this, we will create a web automation:

- We will use selenium
- Important: download the webdriver

```python
from selenium import webdriver #Controlar navegador
from selenium.webdriver.common.keys import Keys #Usar teclado
from selenium.webdriver.common.by import By #Localizar intes no navegador

# Opening a browser
browser = webdriver.Chrome()

# 1 - Get dollar scavenging
browser.get("https://www.google.com/")

browser.find_element(By.XPATH, '/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input').send_keys('translation from dollar to real')

browser.find_element(By.XPATH, '/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input').send_keys(Keys.ENTER)

dollar_scavenging = browser.find_element(By.XPATH, '//*[@id="knowledge-currency__updatable-data-column"]/div[1]/div[2]/span[1]').get_attribute('data-value')

print(dollar_scavenging)

# 2 - Get euro scavenging
browser.get("https://www.google.com/")

browser.find_element(By.XPATH, '/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input').send_keys('translation from euro to real')

browser.find_element(By.XPATH, '/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input').send_keys(Keys.ENTER)

euro_scavenging = browser.find_element(By.XPATH, '//*[@id="knowledge-currency__updatable-data-column"]/div[1]/div[2]/span[1]').get_attribute('data-value')

print(euro_scavenging)

# 3 - Get gold scavenging
browser.get("https://www.melhorcambio.com/ouro-hoje")

gold_scavenging = browser.find_element(By.XPATH, '//*[@id="comercial"]').get_attribute('value')

gold_scavenging = gold_scavenging.replace(",", ".")

print(gold_scavenging)

browser.quit()
```

### Now let's update our price base with the new quotes

- Import Data base

```python
#4 - Import the product list

import pandas as pd

chart = pd.read_excel("Product.xlsx")

display(chart)
```

- Updating Prices and Final Price Calculation

```python
# 5 - Recalculate the price of each product
    # Update the quotation

chart.loc[chart["Coin"]=="Dollar", "Quote"] = float(dollar_scavenging)
chart.loc[chart["Coin"]=="Euro", "Quote"] = float(euro_scavenging)
chart.loc[chart["Coin"]=="Gold", "Quote"] = float(gold_scavenging)

    # update the actual base price (original base price * quote)
chart["Purchase Price"] = chart["Original Price"] * chart["Quote"]

# update the final price (actual base price * Profit)
chart["Sales Price"] = chart["Purchase Price"] * chart["Profit"]

# chart["Sales Price"] = chart["Sales Price"].map("R${:.2f}".format)
display(chart)
```

### Now let's export the new updated price base

```python
# 6 - Save the new product prices
chart.to_excel("NewProduct.xlsx", index=False)
```
